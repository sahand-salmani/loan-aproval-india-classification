```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('loan_approval_dataset.json')
df_copy = df.copy()
```

```python
df.shape
```

| Column Name       | Description                                            |
|-------------------|--------------------------------------------------------|
| Id                | Unique identifier for each loan applicant.             |
| Income            | The income level of the applicant.                     |
| Age               | Age of the applicant.                                  |
| Experience        | Years of professional experience.                     |
| Married/Single    | Marital status of the applicant.                      |
| House_Ownership   | Indicates whether the applicant owns or rents a house.|
| Car_Ownership     | Indicates whether the applicant owns a car.           |
| Profession        | Occupation or profession of the applicant.            |
| CITY              | City of residence of the applicant.                   |
| STATE             | State of residence of the applicant.                  |
| CURRENT_JOB_YRS   | Duration of employment in the current job.            |
| CURRENT_HOUSE_YRS | Duration of residence in the current house.           |
| Risk_Flag         | Indicator of loan risk, where 1 represents a flagged risky applicant and 0 represents a non-risky applicant. |


```python
df.head(3)
```

```python
risk_flag_city_gb = df.groupby(['Risk_Flag', 'CITY']).size().reset_index(name='Count')
risk_flag_city_gb['Rank'] = risk_flag_city_gb.groupby('Risk_Flag')['Count'].rank(ascending=False, method='first')
top_10_cities = risk_flag_city_gb[risk_flag_city_gb['Rank'] <= 10]
pivot_df = top_10_cities.pivot(index='CITY', columns='Risk_Flag', values='Count').fillna(0)


plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
pivot_df.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.legend(title='Risk_Flag', loc='upper right')
plt.title('Counts of Loan Approval by CITY (Top 10)')
plt.xlabel('CITY')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
